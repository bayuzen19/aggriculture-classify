# Skewness Check

If skewness is less than -1 or greater than 1, the distribution is highly skewed.
If skewness is between -1 and -0.5 or between 0.5 and 1, the distribution is moderately skewed.
If skewness is between -0.5 and 0.5, the distribution is approximately symmetric.

```python
#clustering simmilar crops
data_num = df[num].copy()
data_num
```

```python
wcss =[]
for i in range(1,15):
    km=KMeans(n_clusters=i,init="k-means++",max_iter=300,n_init=10,random_state=42)
    km.fit(data_num)
    wcss.append(km.inertia_)

#plotting
plt.figure(figsize=(10,8),dpi=100)
plt.plot(range(1,15),wcss,marker="o",linestyle="--")
plt.title("The Elbow Methode",fontsize=14)
plt.axvline(x=4,color='r',linestyle="--")
plt.xlabel("Number of Clusters",fontsize=12)
plt.ylabel("wccs",fontsize=12)
plt.show()
    
```

```python
km=KMeans(n_clusters=4,init="k-means++",max_iter=300,n_init=10,random_state=42)
y_mean = km.fit_predict(data_num)
y_mean = pd.DataFrame(y_mean)
y_mean.columns =  ['cluster']
cluster_data= df.copy()
cluster_data = pd.concat([cluster_data,y_mean],axis=1)
cluster_data
```
